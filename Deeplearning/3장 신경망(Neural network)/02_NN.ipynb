```python
import numpy as np
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
def identity_function(x):
  return x
```

```python
def softmax(a):
    c = np.max(a)             # list a    
    exp_a = np.exp(a-c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

```python
#  
def init_network():
  network = {}
  network['W1'] = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])       # (2, 3)
  network['b1'] = np.array([[0.1, 0.2, 0.3]])                        # (1, 3)
  network['W2'] = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]])     # (3, 2)
  network['b2'] = np.array([[0.1, 0.2]])                             # (1, 2)
  network['W3'] = np.array([[0.1, 0.3], [0.2, 0.4]])                 # (2, 2)
  network['b3'] = np.array([[0.1, 0.2]])                             # (1, 2)

  return network
```

```python
def forward(network, x):
    W1, W2, W3 = network['W1'], network['W2'], network['W3'] #  3 layer 
    b1, b2, b3 = network['b1'], network['b2'], network['b3']

    a1 = np.dot(x, W1) + b1     # (1, 2) dot (2, 3)  => (1, 3) + (1, 3) = sigmoid
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2) + b2    # (1, 3) dot (3, 2)  => (1, 2) + (1, 2) = sigmoid
    z2 = sigmoid(a2)
    a3 = np.dot(z2, W3) + b3    # (1, 2) dot (2, 2)  => (1, 2) + (1, 2) = identity_function or softmax
    #y = identity_function(a3)
    y = softmax(a3)

    return y
```

```python
network = init_network()
#print(W1.shape)
#print(W2.shape)
#print(W3.shape)

x = np.array([[1.0, 0.5]])
y = forward(network, x)
print(y)

# identity_function [[0.31682708 0.69627909]]
# softmax           [[0.40625907 0.59374093]]
```
