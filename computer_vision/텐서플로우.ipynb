### 텐서플로우 선형 추정

Dense(
  output = 1
  input_dim = 1
  activation = 'linear'
)

  optimizer = 'sgd'
  loss='mse'

</pre>

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([2,4,6,8])
y = np.array([81,93,91,97])

model = Sequential()
```

```python
#모델의 구조(레이어 정의)
model.add(Dense(1, input_dim=1,activation = 'linear')) #Dense(출력되는값,입력될 변수,활성화함수=선형회귀/시그모이드)
#학습 처리방식 정의
model.compile(optimizer='sgd',loss='mse') #optimizer=경사하강법,loss_function=평균제곱오차
model.fit(x,y,epochs=2000)
```

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x),'r')
plt.show()
```

```python
hour = np.array([7])  # reshape 제거
prediction = model.predict(hour)
# print(f"{hour[0]}시간을 공부할 경우의 예상 점수는 {prediction[0][0]:.02f}점입니다.")
print("%.f 시간을 공부할 경우 예상점수는 %.02f점입니다."%(hour,prediction))
```

### 텐서플로우 다중선형회귀
<pre>
Dense(
  output = 1,
  input_dim = 2,
  activation=  'linear'
)
compile(
  optimizer='sgd',
  loss='mse'
)
</pre>

```python
x = np.array([[2,0],[4,4],[6,2],[8,3]])
y = np.array([81,93,91,97])
model = Sequential()

model.add(Dense(1,input_dim=2,activation='linear'))
model.compile(optimizer='sgd',loss='mse')
model.fit(x,y,epochs=2000)
```

```python
# hour = np.array([7])
# private_class = np.array([4])
# prediction = model.predict([[hour,private_class]])
hour = 7
private_class = 4

# 모델에 입력하여 예측합니다.
prediction = model.predict(np.array([[hour, private_class]]))
print("%.f시간을 공부하고 %.f시간의 과외를 받을 경우 예상점수는 %.02f점입니다."%(hour,private_class,prediction))
```
