##기본적인 HTML 문서 구조



- 주요 부분 설명


```
<!DOCTYPE html>:
문서 타입을 정의하는 선언입니다. HTML5 문서임을 나타냅니다.

<html lang="en">:
HTML 문서의 시작을 나타냅니다. lang 속성은 문서의 언어를 지정하며, 여기서는 영어(en)로 설정되어 있습니다.

<head>:
문서의 메타데이터를 담고 있는 섹션입니다.

<meta charset="UTF-8">: 문서의 문자 인코딩을 UTF-8로 설정합니다.

<meta name="viewport" content="width=device-width, initial-scale=1.0">: 모바일 장치에서 화면 크기를 조절하도록 지원합니다.

<title>: 문서의 제목을 정의합니다. 브라우저의 제목 표시줄이나 탭에 표시됩니다.

<body>:
문서의 실제 내용이 들어가는 부분입니다.
텍스트, 이미지, 링크 등이 이곳에 위치합니다.



이러한 HTML 문서 구조는 웹 브라우저가 웹 페이지를 제대로 해석하고 표시할 수 있도록 필요한 최소한의 구조를 제공합니다.
개발자는 <head>와 <body> 안에 필요한 다양한 요소들을 추가하여 웹 페이지를 구성할 수 있습니다.
```

````python
from flask import Flask, request, render_template_string

app = Flask(__name__)

TEMPLATE = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>문서 제목</title>
</head>
<body>

    <!-- 문서 내용이 여기에 들어갑니다. -->
    <header>
        <h1>나의 첫 HTML 페이지</h1>
    </header>

    <main>
        <h3>문서의 실제 내용이 들어가는 부분입니다.</h3>
        <h3>텍스트, 이미지, 링크 등이 이곳에 위치합니다.</h3>
    </main>
</body>
</html>
"""

@app.route('/')
def render_formula():
    return render_template_string(TEMPLATE)

if __name__ == '__main__':
    public_url = ngrok.connect(5000)
    print(f"ngrok URL: {public_url}")
    app.run()
    #app.run(debug=True)
````

##HTML 기초예제 - 본문



```
기본적인 HTML 문서 구조를 보여줍니다.

```



 - 기본적인 HTML 문서 구조를 사용하여 헤더, 내비게이션, 본문, 푸터 등을 포함하고 있습니다.

 - 각 섹션은 링크를 통해 내비게이션에서 이동할 수 있습니다.

- 이 예제를 사용하여 HTML의 기본 구조 및 요소들을 익힐 수 있습니다.

````python
from flask import Flask, request, render_template_string

app = Flask(__name__)

TEMPLATE = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>HTML 기초 예제</title>
</head>
<body>

    <header>
        <h1>나의 첫 HTML 페이지</h1>
    </header>

    <nav>
        <ul>
            <li><a href="#section1">섹션 1</a></li>
            <li><a href="#section2">섹션 2</a></li>
            <li><a href="#section3">섹션 3</a></li>
        </ul>
    </nav>

    <main>
        <section id="section1">
            <h2>섹션 1: 안녕하세요!</h2>
            <p>이것은 첫 번째 섹션입니다. 여기에는 간단한 인사말이 포함되어 있습니다.</p>
        </section>

        <section id="section2">
            <h2>섹션 2: HTML 기초</h2>
            <p>HTML은 HyperText Markup Language의 약자로, 웹 페이지의 구조를 정의하는 언어입니다.</p>
        </section>

        <section id="section3">
            <h2>섹션 3: 간단한 리스트</h2>
            <ul>
                <li>항목 1</li>
                <li>항목 2</li>
                <li>항목 3</li>
            </ul>
        </section>
    </main>

    <footer>
        <p>&copy; 2024 나의 웹사이트</p>
    </footer>

</body>
</html>
"""

@app.route('/')
def render_formula():
    return render_template_string(TEMPLATE)

if __name__ == '__main__':
    public_url = ngrok.connect(5000)
    print(f"ngrok URL: {public_url}")
    app.run()
    #app.run(debug=True)
````

- 본문은 아래와 같이 구분합니다.
  - 헤더(header), 내비게이션(nav), 본문(main), 푸터(footer)는 HTML 문서를 구조화하고 웹 페이지의 각 부분을 식별하기 위한 중요한 요소들입니다.


```
헤더 (<header>):

웹 페이지의 상단 부분을 나타냅니다.
주로 웹 페이지의 제목, 로고, 검색 창, 로그인 링크 등과 같이 핵심적인 내용이 위치합니다.
페이지 상단에 한 번만 사용되며, 전반적인 페이지의 소개와 내용을 담고 있습니다.

예제: html code
<header>
    <h1>나의 웹 페이지</h1>
    <p>웹 페이지의 간단한 설명이 여기에 들어갑니다.</p>
</header>

내비게이션 (<nav>):

웹 페이지의 주요 네비게이션 메뉴를 정의합니다.
일반적으로 메뉴 항목들을 목록(<ul>, <ol>)과 링크(<a>)로 구성합니다.
모바일 반응형 디자인에서는 내비게이션을 숨기고 햄버거 메뉴 등을 활용하기도 합니다.

예제: html code
<nav>
    <ul>
        <li><a href="#home">홈</a></li>
        <li><a href="#about">소개</a></li>
        <li><a href="#contact">문의</a></li>
    </ul>
</nav>

본문 (<main>):

웹 페이지의 주요 내용이 들어가는 부분입니다.
하나의 페이지에는 하나의 <main> 요소만 사용해야 합니다.
헤딩 요소(<h1> ~ <h6>), 단락(<p>), 이미지(<img>), 표(<table>), 폼(<form>) 등이 주로 위치합니다.

예제: html code
<main>
    <h2>최신 소식</h2>
    <p>웹 페이지의 최신 소식이 여기에 나타납니다.</p>
    <!-- 추가적인 내용이 들어갈 수 있음 -->
</main>

푸터 (<footer>):

웹 페이지의 하단 부분을 나타냅니다.
주로 저작권 정보, 연락처 정보, 관련 링크 등이 위치합니다.
페이지 하단에 한 번만 사용되며, 페이지의 마무리와 부가 정보를 담고 있습니다.

예제: html code
<footer>
    <p>&copy; 2024 나의 웹 페이지. 모든 권리 보유.</p>
    <p>문의: contact@example.com</p>
</footer>


이렇게 구조화된 HTML 요소들을 적절하게 조합하여 웹 페이지를 만들면 사용자들이 쉽게 이해하고 탐색할 수 있는 웹 경험을 제공할 수 있습니다.
```


##section1을 누르면 실행하는 예제


```
HTML에서 특정 섹션을 누르면 실행되는 예제를 보여드리겠습니다.

이 예제에서는 각 섹션에 해당하는 내용을 보여주기 위해 JavaScript를 사용합니다.

JavaScript를 사용하여 섹션을 클릭할 때 해당 섹션의 내용이 보이도록 만들어보겠습니다.

```





```
이 예제에서는 showSection이라는 JavaScript 함수를 사용하여 클릭한 섹션만을 화면에 표시합니다.

클릭한 섹션을 보여주기 위해 CSS를 사용하여 초기에는 모든 섹션을 숨김 처리했습니다.

함수가 호출되면 클릭한 섹션만 화면에 표시됩니다.
```




````python
from flask import Flask, request, render_template_string

app = Flask(__name__)

TEMPLATE = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>HTML 섹션 클릭 예제</title>
    <style>
        /* 숨겨진 섹션은 display: none;으로 초기화 */
        .hidden-section {
            display: none;
        }
    </style>
</head>
<body>

    <header>
        <h1>HTML 섹션 클릭 예제</h1>
    </header>

    <nav>
        <ul>
            <li><a href="#section1" onclick="showSection('section1')">섹션 1</a></li>
            <li><a href="#section2" onclick="showSection('section2')">섹션 2</a></li>
            <li><a href="#section3" onclick="showSection('section3')">섹션 3</a></li>
        </ul>
    </nav>

    <main>
        <section id="section1" class="hidden-section">
            <h2>섹션 1: 안녕하세요!</h2>
            <p>이것은 첫 번째 섹션입니다. 여기에는 간단한 인사말이 포함되어 있습니다.</p>
        </section>

        <section id="section2" class="hidden-section">
            <h2>섹션 2: HTML 기초</h2>
            <p>HTML은 HyperText Markup Language의 약자로, 웹 페이지의 구조를 정의하는 언어입니다.</p>
        </section>

        <section id="section3" class="hidden-section">
            <h2>섹션 3: 간단한 리스트</h2>
            <ul>
                <li>항목 1</li>
                <li>항목 2</li>
                <li>항목 3</li>
            </ul>
        </section>
    </main>

    <footer>
        <p>&copy; 2024 나의 웹사이트</p>
    </footer>

    <script>
        // 클릭한 섹션을 보여주는 함수
        function showSection(sectionId) {
            // 모든 섹션을 숨김
            var sections = document.querySelectorAll('.hidden-section');
            sections.forEach(function(section) {
                section.style.display = 'none';
            });

            // 클릭한 섹션만 보임
            var clickedSection = document.getElementById(sectionId);
            clickedSection.style.display = 'block';
        }
    </script>

</body>
</html>
"""

@app.route('/')
def render_formula():
    return render_template_string(TEMPLATE)

if __name__ == '__main__':
    public_url = ngrok.connect(5000)
    print(f"ngrok URL: {public_url}")
    app.run()
    #app.run(debug=True)
````



```
HTML과 JavaScript을 사용하여 섹션을 클릭하면 해당 섹션을 보여주는 코드를 자세히 설명하겠습니다.
이 예제에서는 onclick 이벤트 핸들러를 사용하여 섹션을 클릭할 때 JavaScript 함수가 호출되도록 설정했습니다.
```



```
코드 설명:

각 섹션에 hidden-section 클래스가 지정되어 있습니다.
이 클래스는 섹션을 숨기는 역할을 합니다.
각 내비게이션 링크(<a>)에 onclick 속성이 있으며, 클릭 시 showSection 함수를 호출합니다.
showSection 함수는 모든 섹션을 숨긴 다음, 클릭한 섹션만 보이도록 설정합니다.
이때, display: none;과 display: block;을 사용하여 섹션의 표시 여부를 조절합니다.
이러한 방식으로 사용자가 내비게이션을 클릭하면 해당 섹션의 내용이 동적으로 표시됩니다.
```

## Flask request

Flask에서 request 객체는 클라이언트(사용자)가 서버로 보내는 HTTP 요청에 대한 정보를 담고 있는 객체입니다.

사용자는 웹 브라우저나 다른 HTTP 클라이언트를 통해 서버에 요청(request)을 보내고, 서버는 이에 대한 응답(response)을 보냅니다.

이 요청에는 다양한 정보가 포함될 수 있으며, Flask의 request 객체를 사용하면 이를 쉽게 접근할 수 있습니다.

request 객체는 Flask 애플리케이션에서 자동으로 생성되며, Flask의 flask 모듈에서 가져올 수 있습니다. 주요 기능으로는 다음과 같은 것들이 있습니다.

- request.method: 요청의 HTTP 메소드(GET, POST, PUT, DELETE 등)를 가져옵니다.
- request.args: URL 쿼리 문자열에서 전달된 GET 파라미터를 가져옵니다.
- request.form: POST 요청의 폼 데이터에 접근합니다.
- request.json: JSON 형태로 전송된 데이터를 파싱하여 가져옵니다.
- request.files: 파일 업로드 데이터를 가져옵니다.
- request.headers: 요청의 HTTP 헤더에 접근합니다.
- request.cookies: 클라이언트가 전송한 쿠키 데이터를 가져옵니다.


### 1. GET 요청에서 URL 파라미터 가져오기

위 코드에서 /greet 엔드포인트로 요청을 보내면, URL에 포함된 name 파라미터의 값을 가져와서 인사말을 반환합니다.

예를 들어, /greet?name=John으로 요청하면 "Hello, John!"을 반환합니다.

````python
from flask import Flask, request

app = Flask(__name__)

@app.route('/greet')
def greet():
    # /greet?name=John 와 같은 URL 요청에서 name 파라미터를 가져옵니다.
    name = request.args.get('name', 'Guest')  # 'name' 파라미터가 없으면 'Guest'를 기본값으로 사용
    return f'Hello, {name}!'

if __name__ == '__main__':
    public_url = ngrok.connect(5000)
    print(f"ngrok URL: {public_url}")
    #app.run(debug=True)
    app.run()
````

````python
%%writefile app.py


from flask import Flask, request, url_for, redirect, Response
from post_data import posts
from pyngrok import ngrok

last_id = posts[-1]['id']
app = Flask(__name__)

# '/' 를 요청하면 홈페이지로 가게 된다.
@app.route('/')
def home():
    # url_for() 에 문자열로 함수 이름을 넣어 주면, 알아서 url을 생성해 준다.
    return 'This is Home page.' \
           f'<p><a href="{url_for("home")}">home</a></p>' \
           f'<p><a href="{url_for("about_me")}">about_me</a></p>' \
           f'<p><a href="{url_for("create")}">create new post</a></p>'

# '/about_me' 를 요청하면 자기소개 페이지로 가게 된다.
@app.route('/about_me')
def about_me():
    return 'Let me introduce myself...'

# '/post/read/(포스트의 id)' 를 요청하면 해당 id에 맞는 포스트를 볼 수 있는 페이지로 가게 된다.
@app.route('/post/read/<int:id>')
def read(id):
    title = ''
    subtitle = ''
    content = ''
    for p in posts:
        if id == p['id']:
            title = p['title']
            subtitle = p['subtitle']
            content = p['content']
    return f'제목 : {title}<br/>부제목 : {subtitle}<br/>내용 : {content}<br>'

# '/post/create' 를 요청하면 포스트를 작성할 수 있는 페이지로 가게 된다.
@app.route('/post/create', methods=['GET', 'POST'])
def create():
    #print(request.method, '1')
    # POST 요청이 들어온다면 서버는 게시물을 작성하는 역할을 수행해야 합니다.
    if request.method == 'POST':
        '''
            1. 서버로부터 넘어온 값을 저장하고,
            2. post_data.py 파일을 열고,
            3. post_data.py 의 posts 리스트에 서버에서 넘어온 값을 추가한 다음,
            4. 변경된 post_data.py 파일을 저장
        '''

        # form 태그의 name 속성으로 넘어온 값 저장하기
        title = request.form['title']
        subtitle = request.form['subtitle']
        content = request.form['content']

        # post_data.py 파일 열기
        data = open('post_data.py', 'a')

        global last_id

        # post_data.py 의 posts 리스트에 서버에서 넘어온 값을 추가하기
        data.write(f"\nposts.append({{ 'id':{last_id + 1}, 'title':'{title}', 'subtitle':'{subtitle}', 'content':'{content}' }})")

        data.close()
        last_id = last_id + 1 # id 자동으로 올라가게끔

        return redirect(url_for('home'))
    #print(request.method, '2')
    print(request)
    # GET 요청 시 HTML 폼 반환
    return '''
    <form action="/post/create" method="POST">
        <p><input type="text" name="title" placeholder="제목을 입력하세요."></p>
        <p><input type="text" name="subtitle" placeholder="부제목을 입력하세요."></p>
        <p><textarea name="content" placeholder="내용을 입력하세요"></textarea></p>
        <p><input type="submit" value="게시물 작성하기"></p>
    </form>
    '''

if __name__ == '__main__':
    public_url = ngrok.connect(5000)
    print(f"ngrok URL: {public_url}")
    #app.run(debug=True)
    app.run()
````
