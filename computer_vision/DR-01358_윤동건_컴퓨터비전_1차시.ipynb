### 행렬생성 매서드

#### np.arange(from,to,step)   
+ from부터 to까지 step을 증가시키며 생성
+ 숫자 하나만 넣으면 0부터 숫자개의 정수 생성
+ 숫자. 넣으면 0부터 숫자계의 실수 생성   

#### np.linspace(from,to,npoints)
+ from에서 to까지 npoint개의 값이 생성되게 분배  

#### np.eye(n): 단위행렬생성


```python
np.arange(1,5,0.1)
```

```python
np.arange(5)
```

```python
np.arange(5.)
```

```python
np.linspace(1,10,5)
```

```python
np.eye(3)
```
