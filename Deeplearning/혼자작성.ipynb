```python
import numpy as np
import matplotlib.pyplot as plt

data = [[2, 81], [4, 93], [6, 91], [8, 97]]
x = [i[0] for i in data]
y = [i[1] for i in data]

plt.figure(figsize = (8,5))
plt.scatter(x,y)
plt.show()
```

```python
x_data = np.array(x)
y_data = np.array(y)

a = 0
b = 0
lr = 0.03
epochs = 2001

for i in range(epochs):
    y_hat = a*x_data+b
    error = y_data - y_hat
    a_diff = -(2/len(x_data)*sum((error)*(x_data)))
    b_diff = -(2/len(x_data)*sum(error))
    a -= lr*a_diff
    b -= lr*b_diff
    if i%100==0:
        print('epochs=%.f, 기울기=%.04f , 절편=%.04f'%(i,a,b))

y_pred = a*x_data + b
plt.scatter(x,y)
plt.plot([min(x_data),max(x_data)],[min(y_pred),max(y_pred)])
plt.show()
    
```

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
import numpy as np
import pandas as pd
import tensorflow as tf


seed=0
np.random.seed(3)
tf.random.set_seed(3)

file_path = '/home/donggeonyoon/딥러닝 파일/080228/deeplearning/dataset/sonar.csv'
df = pd.read_csv(file_path,header=None)
# print(df.head())
# print(df.info())

dataset = df.values
X = dataset[:,0:60].astype(float)
Y_obj = dataset[:,60]

e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)

n_fold=10
skf = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=seed)
accuracy=[]

for train,test in skf.split(X,Y):
    model = Sequential()
    model.add(Dense(24,  input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(loss='mean_squared_error',
                optimizer='adam',
                metrics=['accuracy'])

    model.fit(X[train],Y[train], epochs=50,batch_size=5)
    k_accuracy = "%.4f" % (model.evaluate(X[test], Y[test])[1])
    accuracy.append(k_accuracy)

print("\n %.f fold accuracy:" % n_fold, accuracy)
```

Mnist-xavier 가중치초기화

```python
import torch
import torchvision.datasets as dsets
import torchvision.transforms as transforms
import random

learning_rate = 0.001
training_epochs = 15
batch_size = 100

mnist_train = dsets.MNIST(root='/home/arclab/딥러닝 데이터/MNIST_data/',
                          train=True,
                          transform=transforms.ToTensor(),
                          download=True)
#60000,1,28,28인데 dsets객체라서 shape못봄
mnist_test = dsets.MNIST(root='/home/arclab/딥러닝 데이터/MNIST_data/',
                          train=False,
                          transform=transforms.ToTensor(),
                          download=True)

data_loader = torch.utils.data.DataLoader(dataset=mnist_train,
                                          batch_size=batch_size,
                                          shuffle=True)

layer1 = torch.nn.Linear(784, 256, bias=True)
layer2 = torch.nn.Linear(256, 256, bias=True)
layer3 = torch.nn.Linear(256, 10, bias=True)
relu = torch.nn.ReLU()

torch.nn.init.xavier_uniform(layer1.weight)
torch.nn.init.xavier_uniform(layer2.weight)
torch.nn.init.xavier_uniform(layer3.weight)

model = torch.nn.Sequential(layer1,relu,layer2,relu,layer3)
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(),lr=learning_rate)

for epoch in range(training_epochs):
    avg_cost = 0
    total_batch = len(mnist_train) // batch_size #60000/100

    for i,(batch_xs, batch_ys) in enumerate(data_loader):
        X = batch_xs.view(-1,28*28)
        Y = batch_ys

        optimizer.zero_grad()
        hypothesis = model(X)
        cost = criterion(hypothesis,Y)
        cost.backward()
        optimizer.step()

        avg_cost += cost/total_batch
        print("[Epoch: {:>4}] cost = {:>.9}".format(epoch + 1, avg_cost.item()))

print('Learning Finished!')

```

```python
X_test = mnist_test.test_data.view(-1,28*28).float()
Y_test = mnist_test.test_labels

prediction = model(X_test)
correct_prediction = (torch.max(prediction,1)[1] == Y_test.data) #각 이미지의 10개 출력값중 가장 높은놈의 레이블 == 결과 레이블
accuracy = correct_prediction.float().mean() #60000개 중에 몇개 맞췄나
print('Accuracy: ',accuracy)

r = random.randint(0, len(mnist_test) - 1)
X_single_data = mnist_test.test_data[r:r+1].view(-1,28*28).float()
Y_single_data = mnist_test.test_labels[r:r+1]#배치차원을 유지하기 위함

print('Label: ',Y_single_data)
single_prediction = model(X_single_data)
print("prediction ", torch.max(single_prediction,1)[1])
```

```python
import torch
import torchvision.datasets as dsets
import torchvision.transforms as transforms
import random

learning_rate = 0.001
training_epochs = 15
batch_size = 100

mnist_train = dsets.MNIST(root='/home/arclab/딥러닝 데이터/MNIST_data/',
                          train=True,
                          transform=transforms.ToTensor(),
                          download=True)
#60000,1,28,28인데 dsets객체라서 shape못봄
mnist_test = dsets.MNIST(root='/home/arclab/딥러닝 데이터/MNIST_data/',
                          train=False,
                          transform=transforms.ToTensor(),
                          download=True)

data_loader = torch.utils.data.DataLoader(dataset=mnist_train,
                                          batch_size=batch_size,
                                          shuffle=True)

layer1 = torch.nn.Linear(784, 512, bias=True)
layer2 = torch.nn.Linear(512, 512, bias=True)
layer3 = torch.nn.Linear(512, 512, bias=True)
layer4 = torch.nn.Linear(512, 512, bias=True)
layer5 = torch.nn.Linear(512, 10, bias=True)
relu = torch.nn.ReLU()

dropout = torch.nn.Dropout(p= 1-0.7)

torch.nn.init.xavier_uniform(layer1.weight)
torch.nn.init.xavier_uniform(layer2.weight)
torch.nn.init.xavier_uniform(layer3.weight)
torch.nn.init.xavier_uniform(layer4.weight)
torch.nn.init.xavier_uniform(layer5.weight)

model = torch.nn.Sequential(layer1,relu,dropout,
                            layer2,relu,dropout,
                            layer3,relu,dropout,
                            layer4,relu,dropout,
                            layer5)

criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(),lr=learning_rate)

for epoch in range(training_epochs):
    avg_cost = 0
    total_batch = len(mnist_train) // batch_size #60000/100

    for i,(batch_xs, batch_ys) in enumerate(data_loader):
        X = batch_xs.view(-1,28*28)
        Y = batch_ys

        optimizer.zero_grad()
        hypothesis = model(X)
        cost = criterion(hypothesis,Y)
        cost.backward()
        optimizer.step()

        avg_cost += cost/total_batch
        print("[Epoch: {:>4}] cost = {:>.9}".format(epoch + 1, avg_cost.item()))

print('Learning Finished!')

```

```python
X_test = mnist_test.test_data.view(-1,28*28).float()
Y_test = mnist_test.test_labels

prediction = model(X_test)
correct_prediction = (torch.max(prediction,1)[1] == Y_test.data) #각 이미지의 10개 출력값중 가장 높은놈의 레이블 == 결과 레이블
accuracy = correct_prediction.float().mean() #60000개 중에 몇개 맞췄나
print('Accuracy: ',accuracy)

r = random.randint(0, len(mnist_test) - 1)
X_single_data = mnist_test.test_data[r:r+1].view(-1,28*28).float()
Y_single_data = mnist_test.test_labels[r:r+1]#배치차원을 유지하기 위함

print('Label: ',Y_single_data)
single_prediction = model(X_single_data)
print("prediction ", torch.max(single_prediction,1)[1])
```

```python
import torch
import torchvision.datasets as dsets
import torchvision.transforms as transforms
import random

learning_rate = 0.001
training_epochs = 15
batch_size = 100

mnist_train = dsets.MNIST(root='/home/arclab/딥러닝 데이터/MNIST_data/',
                          train=True,
                          transform=transforms.ToTensor(),
                          download=True)
#60000,1,28,28인데 dsets객체라서 shape못봄
mnist_test = dsets.MNIST(root='/home/arclab/딥러닝 데이터/MNIST_data/',
                          train=False,
                          transform=transforms.ToTensor(),
                          download=True)

data_loader = torch.utils.data.DataLoader(dataset=mnist_train,
                                          batch_size=batch_size,
                                          shuffle=True)
```

X → __call__(X) → forward(X) → 순전파 연산 → output     
[60000,1,28,28] -- (batch) -> [100,1,28,28] --(Conv2d)-> [100,32,28,28] --(Pooling)-> [100,32,14,14]
--(Conv2d)-> [100,64,14,14] --(Pooling)-> [100,64,7,7] --(FCN)



$$
\text{출력 크기} = \left\lfloor \dfrac{\text{입력 크기} + 2 \times \text{패딩} - (\text{커널 크기} - 1) - 1}{\text{스트라이드}} \right\rfloor + 1
$$

super().__init__()은 상속한 클래스의 매소드를 호출 즉 init을 호출해 속성을 확실히 초기화하는것
- 부모 클래스의 속성과 매서드 초기화

```python
class CNN(torch.nn.Module):

    def __init__(self):
        super().__init__()
        self.layer1 = torch.nn.Sequential(
            torch.nn.Conv2d(1, 32, kernel_size=3, stride=1, padding=1),
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size=2, stride=2))
        
        self.layer2 = torch.nn.Sequential(
            torch.nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1),
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size=2, stride=2))
        
        self.fc = torch.nn.Linear(64*7*7,10,bias=True)
        torch.nn.init.xavier_uniform(self.fc.weight)
    
    def forward(self, x):
        x1 = self.layer1(x)
        x2 = self.layer2(x1)
        x2_flat = x2.view(x2.size(0),-1) #out = [100,64,7,7]인데 이를 [100,64*7*7]로
        out = self.fc(x2_flat)
        return out
```

```python
model = CNN()
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(),lr=learning_rate)

print('Learning started!')
for epoch in range(training_epochs):
    avg_cost = 0
    total_batch = len(mnist_train) // batch_size #60000/100

    for i,(batch_xs, batch_ys) in enumerate(data_loader):
        X = batch_xs
        Y = batch_ys

        optimizer.zero_grad()
        hypothesis = model(X)
        cost = criterion(hypothesis,Y)
        cost.backward()
        optimizer.step()

        avg_cost += cost/total_batch
        print("[Epoch: {:>4}] cost = {:>.9}".format(epoch + 1, avg_cost.item()))

print('Learning Finished!')
```

```python
import random  # random 모듈 임포트

# 테스트 데이터 준비
X_test = mnist_test.data.float()
Y_test = mnist_test.targets

X_test = X_test.view(-1, 1, 28, 28)

# 모델 예측 및 정확도 계산
model.eval()
with torch.no_grad():
    prediction = model(X_test)
    correct_prediction = (torch.max(prediction, 1)[1] == Y_test)
    accuracy = correct_prediction.float().mean()
    print('Accuracy: ', accuracy.item())

# 랜덤한 테스트 샘플 선택 및 예측
r = random.randint(0, len(mnist_test) - 1)
X_single_data = mnist_test.data[r:r+1].float()
Y_single_data = mnist_test.targets[r:r+1]

X_single_data = X_single_data.view(-1, 1, 28, 28)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
# cuda 사용 시 데이터 이동
X_single_data = X_single_data.to(device)
Y_single_data = Y_single_data.to(device)

print('Label: ', Y_single_data.item())

model.eval()
with torch.no_grad():
    single_prediction = model(X_single_data)
    predicted_label = torch.argmax(single_prediction, 1)
    print("Prediction: ", predicted_label.item())
```

mnist-deep-cnn

```python
import torch
import torchvision.datasets as dsets
import torchvision.transforms as transforms
import random

learning_rate = 0.001
training_epochs = 15
batch_size = 100
keep_prob = 0.7

mnist_train = dsets.MNIST(root='/home/arclab/딥러닝 데이터/MNIST_data/',
                          train=True,
                          transform=transforms.ToTensor(),
                          download=True)
#60000,1,28,28인데 dsets객체라서 shape못봄
mnist_test = dsets.MNIST(root='/home/arclab/딥러닝 데이터/MNIST_data/',
                          train=False,
                          transform=transforms.ToTensor(),
                          download=True)

data_loader = torch.utils.data.DataLoader(dataset=mnist_train,
                                          batch_size=batch_size,
                                          shuffle=True)

class CNN(torch.nn.Module):

    def __init__(self):
        super().__init__()
        #[100,1,28,28]->[100,32,14,14]
        self.layer1 = torch.nn.Sequential(
            torch.nn.Conv2d(1, 32, kernel_size=3, stride=1, padding=1), 
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size=2, stride=2),
            torch.nn.Dropout(p=1-keep_prob))
            
        #[100,32,14,14]->[100,64,7,7]
        self.layer2 = torch.nn.Sequential(
            torch.nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1),
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size=2, stride=2),
            torch.nn.Dropout(p=1-keep_prob))
        #->[100,128,3,3] or [100,128,4,4](ceil)
        self.layer3 = torch.nn.Sequential(
            torch.nn.Conv2d(64,128,kernel_size=3,stride=1,padding=1),
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size=2, stride=2,ceil_mode=True), #3.5에서 반올림 or 패딩추가
            torch.nn.Dropout(p=1-keep_prob))
        
        self.fc1 = torch.nn.Linear(128*4*4,625,bias=True)
        torch.nn.init.xavier_uniform(self.fc1.weight)

        self.layer4 = torch.nn.Sequential(
        self.fc1,
        torch.nn.ReLU(),
        torch.nn.Dropout(p=1 - keep_prob))

        self.fc2 = torch.nn.Linear(625, 10, bias=True)
        torch.nn.init.xavier_uniform(self.fc2.weight)    
    
    def forward(self, x):
        x1 = self.layer1(x)
        x2 = self.layer2(x1)
        x3 = self.layer3(x2)
        x3_flat = x3.view(x3.size(0),-1) #out = [100,64,7,7]인데 이를 [100,64*7*7]로
        out1 = self.fc1(x3_flat)
        out2 = self.fc2(out1)
        return out2

model = CNN()
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(),lr=learning_rate)

print('Learning started!')
for epoch in range(training_epochs):
    avg_cost = 0
    total_batch = len(mnist_train) // batch_size #60000/100

    for i,(batch_xs, batch_ys) in enumerate(data_loader):
        X = batch_xs
        Y = batch_ys

        optimizer.zero_grad()
        hypothesis = model(X)
        cost = criterion(hypothesis,Y)
        cost.backward()
        optimizer.step()

        avg_cost += cost/total_batch
        print("[Epoch: {:>4}] cost = {:>.9}".format(epoch + 1, avg_cost.item()))

print('Learning Finished!')
```

```python
import random  # random 모듈 임포트

# 테스트 데이터 준비
X_test = mnist_test.data.float()
Y_test = mnist_test.targets

X_test = X_test.view(-1, 1, 28, 28)

# 모델 예측 및 정확도 계산
model.eval()
with torch.no_grad():
    prediction = model(X_test)
    correct_prediction = (torch.max(prediction, 1)[1] == Y_test)
    accuracy = correct_prediction.float().mean()
    print('Accuracy: ', accuracy.item())

# 랜덤한 테스트 샘플 선택 및 예측
r = random.randint(0, len(mnist_test) - 1)
X_single_data = mnist_test.data[r:r+1].float()
Y_single_data = mnist_test.targets[r:r+1]

X_single_data = X_single_data.view(-1, 1, 28, 28)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
# cuda 사용 시 데이터 이동
X_single_data = X_single_data.to(device)
Y_single_data = Y_single_data.to(device)

print('Label: ', Y_single_data.item())

model.eval()
with torch.no_grad():
    single_prediction = model(X_single_data)
    predicted_label = torch.argmax(single_prediction, 1)
    print("Prediction: ", predicted_label.item())
```
