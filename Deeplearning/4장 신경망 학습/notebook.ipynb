```python
def function_1(x):
    return 0.01*x**2 + 0.1*x

print(function_1(5))
print(function_1(10))
```

```python
import matplotlib.pylab as plt
import numpy as np

x = np.arange(0, 20, 1)
y = function_1(x)
print(y)
print(y.min(),y.max())
# plt.ylim(0.1, 6)
plt.plot(x,y)
plt.show()
```

```python
def numerical_diff(f,x):
    h = 10e-4
    return (f(x + h)-f(x - h))/(2*h)
```

```python
x_tangent = np.arange(0, 20, 1)
tangent_slope = numerical_diff(function_1,10)
y_at_10 = function_1(10)
y_tangent = tangent_slope*(x-10)+y_at_10

print(y_at_10)
print(function_1(10))

plt.plot(x,y)
plt.plot(x_tangent,y_tangent,color='red')
plt.show()
```
