```python
import numpy as np
import matplotlib.pyplot as plt

x_data = np.array([1,2,3])
y_data = np.array([2,4,6])
```

```python
w = 1
def forward(x):
    return x*w

def loss(x,y):
    y_pred = forward(x)
    return (y_pred - y)**2
```

```python
w_list = []
mse_list = []

for w in np.arange(0.0,4.1,0.1):
    print('w=',w)
    l_sum = 0
    for x_val,y_val in zip(x_data,y_data):
        y_pred_val = forward(x_val)
        l = loss(x_val,y_val)
        l_sum += l
        print('\t',x_val,y_val,y_pred_val,l)
    print('MSE',l_sum / len(x_data))
    w_list.append(w)
    mse_list.append(l_sum/len(x_data))
```

```python
plt.plot(w_list,mse_list,label='MSE')
plt.ylabel('Loss')
plt.xlabel('W')
plt.legend()
plt.show()
```

```python
x_data = [1.0,2.0,3.0]
y_data = [2.0,4,0,6.0]

w = 1.0
def forward(x):
    return x*w

def loss(x,y):
    y_pred = forward(x)
    return (y_pred - y)**2

def gradiant(x,y):
    return 2*x*(x*w-y)
```

```python
#befire training
print('predict(before training)',4,forward(4))

for epoch in range(100):
    for x_val,y_val in zip(x_data,y_data):
        grad = gradiant(x_val,y_val)
        w = w-0.01 * grad
        print("\tgrad: ",x_val,y_val,grad)
        l = loss(x_val,y_val)

    print(f'progress: {epoch} w = {w} loss = {l}')


print('predict(atfer training) ',forward(4))        
```
