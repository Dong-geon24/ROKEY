### Mnist with MLP_scratch

```python
import torch
import torchvision.datasets as dsets
import torchvision.transforms as transforms
import random

learning_rate = 0.001
training_epochs = 15
batch_size = 100

mnist_train = dsets.MNIST(root='/home/arclab/딥러닝_데이터/dataset/MNIST_data',
                          train=True,
                          transform=transforms.ToTensor(),
                          download=True)
mnist_test = dsets.MNIST(root='/home/arclab/딥러닝_데이터/dataset/MNIST_data',
                          train=False,
                          transform=transforms.ToTensor(),
                          download=True)
data_loader = torch.utils.data.DataLoader(dataset=mnist_train,
                                          batch_size=batch_size,
                                          shuffle=True)
#[100,1,28,28]
```

```python
layer1 = torch.nn.Linear(784, 512, bias=True)
layer2 = torch.nn.Linear(512, 512, bias=True)
layer3 = torch.nn.Linear(512, 512, bias=True)
layer4 = torch.nn.Linear(512, 512, bias=True)
layer5 = torch.nn.Linear(512, 10, bias=True)
relu = torch.nn.ReLU()

model = torch.nn.Sequential(layer1,relu,
                            layer2,relu,
                            layer3,relu,
                            layer4,relu,
                            layer5)
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(),lr=learning_rate)
```

```python
for epoch in range(training_epochs):
    avg_cost=0
    total_batch = len(mnist_train)//batch_size

    for i,(batch_xs,batch_ys) in enumerate(data_loader):
        X = batch_xs.view(-1,28*28).float() #100,784
        Y = batch_ys

        optimizer.zero_grad()
        hypothesis = model(X)
        cost = criterion(hypothesis,Y)
        cost.backward()
        optimizer.step()

        avg_cost += cost/total_batch
        print("[Epoch: {:>4}] cost = {:>.9}".format(epoch + 1, avg_cost.item()))
print('learning finished!')
```

```python
X_test = mnist_test.test_data.view(-1,28*28).float()
Y_test = mnist_test.test_labels

prediction = model(X_test)
correct_prediction = (torch.max(prediction,1)[1] == Y_test.data)
accuracy = correct_prediction.float().mean()
print('Accuracy: ',accuracy)

r = random.randint(0, len(mnist_test)-1)
X_single_data = mnist_test.test_data[r:r+1].view(-1,28*28).float()
Y_single_data = mnist_test.test_labels[r:r+1]

print('Label: ',Y_single_data)
single_prediction = model(X_single_data)
print('Prediction :',torch.max(single_prediction,1)[1])
```

```python
from tensorflow import keras
from tensorflow.keras import layers
inputs = keras.Input(shape=(28,28,1))
x = layers.Conv2D(filters=32, kernel_size=3, activation="relu")(inputs)
x = layers.MaxPooling2D(pool_size=2)(x)
x = layers.Conv2D(filters=64, kernel_size=3, activation="relu")(x)
x = layers.MaxPooling2D(pool_size=2)(x)
x = layers.Conv2D(filters=128, kernel_size=3, activation="relu")(x)
x = layers.Flatten()(x)
outputs = layers.Dense(10, activation="softmax")(x)
model = keras.Model(inputs=inputs, outputs=outputs)

model.summary()
```

```python
from tensorflow.keras.datasets import mnist

(train_images, train_labels),(test_images,test_labels) = mnist.load_data()
train_images = train_images.reshape((60000,28,28,1)).astype('float32') / 255
test_images = test_images.reshape((10000,28,28,1)).astype('float32') / 255

model.compile(optimizer='rmsprop',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(train_images, train_labels, epochs=5, batch_size=64)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
print(f"테스트정확도: {test_acc:.3f}")
```
